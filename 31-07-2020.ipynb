```python
import matplotlib.pyplot as pt
xaxis=[25,26,27,28]
yaxis=[2800,2900,3000,500]
pt.plot(xaxis,yaxis,color='k',linestyle='--',marker='o',linewidth=2)
y1axis=[2000,4000,9000,1000]
pt.plot(xaxis,y1axis,color='#5a7d9a',marker='d',label='yaxis')
pt.title('basic plot')
pt.xlabel('ages')
pt.ylabel('salary')
pt.show()
```
