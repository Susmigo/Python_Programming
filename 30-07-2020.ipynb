```python
import pandas as pd
```

```python
# merge
df = pd.DataFrame({"weight":[10,20,30,40],"num":[50,60,70,80],"age":[1,2,3,4]},index=[2001,2002,2003,2004])
df
```

```python
df1 = pd.DataFrame({"weight":[10,20,30,40],"num":[50,60,70,80],"age":[1,2,3,4]},index=[2001,2002,2003,2004])
df1
```

```python
df1.merge(df)
```

```python
df2 = pd.DataFrame({"weight":[5,4,3,8],"num":[47,87,57,64],"age":[15,25,35,45]},index=[2001,2002,2003,2004])
df2
```

```python
df1.merge(df2)
```

```python
# Join
df4 = pd.DataFrame({"weight":[5,4,3,8],"num":[47,87,57,64],"age":[15,25,35,45]},index=[2001,2002,2003,2004])
df4
```

```python
df1.join(df4)
```

```python
df6 = pd.DataFrame({"weight1":[5,4,3,8],"num1":[47,87,57,64],"age1":[15,25,35,45]},index=[2001,2002,2003,2004])
df6
```

```python
df1.join(df6)
```

```python
# concatenation
c=pd.concat([df1,df6])
c
```

```python
# read csv file data
filepath='Datafiles/income.csv'
import pandas as pd
def readCsvdata(filepath):
    return pd.read_csv(filepath)
readCsvdata(filepath)
```

```python
# to print all columns in a single line
incomedt=readCsvdata(filepath)
def printdataframecols(df):
    columns=df.columns
    for column in columns:
        print(column,end=" ")
    return
printdataframecols(incomedt)
```

```python
# Average income of all states from 2005-2013
def avgincome(df):
    for i in range(0,5):
        columns=df.values[i][2:]
        total=0
        count=0
        for column in columns:
            total=total+column
            count=count+1
        print(df.values[i][1],total//count)
    return
incomedt=readCsvdata(filepath)
avgincome(incomedt)
```

```python
import matplotlib.pyplot as plt
plt.title('linear graph')
plt.plot([1,2,3],[4,5,6])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Non linear graph
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v,'ro')
plt.title('non linear ')
plt.xlabel('x-axis(time)')
plt.ylabel('y-axis(velocity)')
plt.grid()
plt.show()
```

```python
# bar graph
x=[1,2,3,4]
y=[10,20,30,40]
plt.bar(x,y)
plt.show()
```

```python
# pie chart
x=[1,2,3,4]
explode = (0, 0.1, 0, 0)
labels=['a','b','c','d']
plt.pie(x,explode=explode,labels=labels)
plt.title('basic pie ')
plt.show()
```
